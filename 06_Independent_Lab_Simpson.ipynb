```python
#Great! Now that we have a general idea of how the program should function, let's get coding!

#First let's start by establishing the various lists and variables required for the functionality of the program 

quit_program = 0 #A variable used to determine whether or not the user has quit the program

navigation_return = 0 #Determines whether or not the user has decided to return to the navigation menu

number_students = 4 #This is to hold the indexing value for the students

students = ["Carmen Berzatto","Sydney Adamu","Neil Fak","Marcus Brooks"] #This is the initial list of students provided

GPA = [0,0,0,0] #Establish the container variable for GPA

Major = ["None", "None","None","None"] #Establish the container variable for Major

while(quit_program == 0): 
    
    navigation_selection = 0 #Reset the navigation selection variable to accept fresh input
    
    print("Welcome to the Navigation Menu!") #This chunk of code is for the navigation menu which will list all of the different options
    print("Please select one of the following:")
    print("1. Student Information Query")
    print("2. Student Information Entry")
    print("3. GPA Metrics Query")
    print("4. Academic Violations")
    print("5. Quit the Program")
    print("")                                #This is the end of the navigation menu
    
    navigation_selection = float(input(""))         #Grabs the user input for the navigation menu
    
    
    #Student Information Query
    if(navigation_selection == 1):  
        print("")
        print("| Name, GPA, Major  |")
        for s in students:    
            indx = students.index(s)
            print(f"|{s}  {GPA[indx]}  {Major[indx]} |")
        choice = input("Please enter Y to return to the navigation menu")
        if(choice == "Y"):
            print("")
            exit()
    
    
    #Student Information Entry
    elif(navigation_selection == 2):  
        
        print("")  #Printing of the intro to the section
        print("Hello and welcome to the student information entry, please select an option.")
        print("")
        print("1. New Student")
        print("2. GPA")
        print("3. Major")
        print("")
        entry_choice = float(input(""))  #Takes the user input for the selection
        if(entry_choice == 1):
            new_student = input("Please enter the student's first and last name")
            if(new_student != ""):
                number_students += 1
                students.insert(number_students, new_student)
                exit()
            else:
                exit()
        elif(entry_choice == 2):  #Handles the GPA Change
            print("")
            print("Please select a student")
            print("")
            for s in students:
                print(students.index(s)," ",s)
            print("")
            student_choice = int(input(""))
            print(students[student_choice])
            print("")
            print("Confirm? (Y for yes)")
            choice = input("")
            if choice == "Y": 
                print("")
                print("Please enter the GPA:")
                print("")
                new_GPA = float(input(""))
                print ("")
                GPA.insert(student_choice, new_GPA)
                print("")
                exit()
            else:
                exit()
                
        elif(entry_choice == 3):  #Handles the changing of the majors
            print("")
            print("Please select a student")
            print("")
            for s in students:
                print(students.index(s), " ",s)
            print("")
            student_choice = int(input(""))
            print(students[student_choice])
            print("")
            print("Confirm? (Y for yes)")
            choice = input("")
            if choice == "Y":
                print("")
                print("Please enter the major:")
                print("")
                new_major = input("")
                print("")
                Major.insert(student_choice, new_major)
                print("")
                exit()
            else:
                exit()
                
    
    elif(navigation_selection == 3):  #Handles the GPA Metrics Query
        maximum_gpa = max(GPA)
        minimum_gpa = min(GPA)
        average_gpa = sum(GPA) / len(GPA)
        print("")
        print(f"The max GPA: {maximum_gpa}")
        print("")
        print(f"The min GPA: {minimum_gpa}")
        print("")
        print(f"The Average GPA: {average_gpa}")
        print("")
        print("Would you like to return? (Y for yes)")
        print("")
        choice = input("")
        print("")
        if choice == "Y":
            exit()
        else:
            exit()
            
            
        
    #elif(navigation_selection == 4):  #Handles the academic violations
    
    
    
    
    
    #elif(navigation_selection == 5):  #Exits the program
        quit()
            
```
