-Student Informataion Entry (Where the end user will be able to input the student's first and last name, major and GPA) <br>

```python
#Great! Now that we have a general idea of how the program should function, let's get coding!

#First let's start by establishing the various lists and variables required for the functionality of the program 



quit_program = 0 #A variable used to determine whether or not the user has quit the program

navigation_return = 0 #Determines whether or not the user has decided to return to the navigation menu

number_students = 4 #This is to hold the indexing value for the students

students = ["Carmen Berzatto","Sydney Adamu","Neil Fak","Marcus Brooks"] #This is the initial list of students provided

GPA = [4,2,3,2] #Establish the container variable for GPA

Major = ["None", "None","None","None"] #Establish the container variable for Major

Student_Violations = ["None","None","None","None"] #Container for the student violations

Violations = ["Copying Homework","Cheating on Exam","Plagiarism","Cheating on Quiz","Writing paper for Another Student"]



#Great, now lets work on getting the program rocking and rolling!

while(quit_program == 0): 
    
    navigation_selection = 0 #Reset the navigation selection variable to accept fresh input
    
    print("Welcome to the Navigation Menu!") #This chunk of code is for the navigation menu which will list all of the different options
    print("Please select one of the following:")
    print("1. Student Information Query")
    print("2. Student Information Entry")
    print("3. GPA Metrics Query")
    print("4. Academic Violations")
    print("5. Quit the Program")
    print("")                                #This is the end of the navigation menu
    
    navigation_selection = float(input(""))         #Grabs the user input for the navigation menu
    
    
    #Student Information Query
    
    if(navigation_selection == 1):  
        print("")
        print("| Name, GPA, Major, Violations  |")
        for s in students:    
            indx = students.index(s)
            print(f"|{s}:  {GPA[indx]},  {Major[indx]},  {Student_Violations[indx]} |")
        choice = input("Please enter Y to return to the navigation menu")
        if(choice == "Y"):
            print("")
            exit()
    
    
    #Student Information Entry
    
    elif(navigation_selection == 2):  
        
        print("")  #Printing of the intro to the section
        print("Hello and welcome to the student information entry, please select an option.")
        print("")
        print("1. New Student")
        print("2. GPA")
        print("3. Major")
        print("")
        entry_choice = float(input(""))  #Takes the user input for the selection
        
        
        #Add a new student
        
        if(entry_choice == 1):
            new_student = input("Please enter the student's first and last name")
            if(new_student != ""):
                number_students += 1
                students.insert(number_students, new_student)
                GPA.insert(number_students, 0)
                Major.insert(number_students, "None")
                Student_Violations.insert(number_students, "None")
                exit()
            else:
                exit()
                
                
        #GPA Change 
        
        elif(entry_choice == 2):  
            print("")
            print("Please select a student")
            print("")
            for s in students:
                print(students.index(s)," ",s)
            print("")
            student_choice = int(input(""))
            print(students[student_choice])
            print("")
            print("Please enter the GPA:")
            print("")
            new_GPA = float(input(""))
            print ("")
            print(f"Student: {students[student_choice]}; GPA: {new_GPA}")
            print("")
            print("Confirm? (Y for yes)")
            choice = input("")
            if choice == "Y": 
                print("")
                GPA.insert(student_choice, new_GPA)
                print("")
                exit()
            else:
                exit()
                
                
                
        #Changing Majors
                
        elif(entry_choice == 3):  
            print("")
            print("Please select a student")
            print("")
            for s in students:
                print(students.index(s), " ",s)
            print("")
            student_choice = int(input(""))
            print("")
            print("Please enter the major:")
            print("")
            new_major = input("")
            print("")
            print(f"Student: {students[student_choice]}; Major: {new_major}")
            print("")
            print("Confirm? (Y for yes)")
            choice = input("")
            if choice == "Y":
                Major.insert(student_choice, new_major)
                print("")
                exit()
            else:
                exit()
                
    
    #GPA Metrics Query
    
    elif(navigation_selection == 3): 
        maximum_gpa = max(GPA)
        minimum_gpa = min(GPA)
        average_gpa = sum(GPA) / len(GPA)
        print("")
        print(f"The max GPA: {maximum_gpa}")
        print("")
        print(f"The min GPA: {minimum_gpa}")
        print("")
        print(f"The Average GPA: {average_gpa}")
        print("")
        print("Would you like to return? (Y for yes)")
        print("")
        choice = input("")
        print("")
        if choice == "Y":
            exit()
        else:
            exit()
            
            
        
    #Academic Violations
        
    elif(navigation_selection == 4):  
        print("")
        print("Please select a student:")
        for s in students:
            print("")
            print(students.index(s), " ",s)  #List all of the students
        print("")
        student_choice = int(input(""))  #Grab the User Input
        print("")
        print("Please select a violation from the list:")
        print("")
        for i in Violations:
            print(Violations.index(i), " ", i)
        print("")
        Violation_Selection = int(input(""))  #Select a Violation
        print("")
        print(f"Student: {students[student_choice]}; Violation: {Violations[Violation_Selection]}") #Print to confirm
        print("")
        print("Confirm? (Y for yes)")
        print("")
        choice = input("")
        print("")
        if choice == "Y":
            Student_Violations.insert(student_choice, Violations[Violation_Selection])  #Add the violation to the student
        else:
            exit()
    
    elif(navigation_selection == 5):  #Exits the program
        quit_program = 1
            
```

# Code Breakdown


## Navigation Menu
The User will have a navigation menu to which they can return. This allows the end user to then be able to "move" around the program and not be stuck with whatever the first choice they selected. This is handled by having the navigation menu always appear first within the While() loop. This while loop allows for the program to continuously run until the end user decides to quit the program. Under the navigation menu, the end user is provided a list of items from which they can select. 

### Navigation Menu Selection
The selection from the user is handled by taking their numeric input, which aligns with one of the few options. Each individual options is then handled by an if() statement. This can be though of as if the user selects "1" then the if() statement handeling that selection will then kick in. 


## Lists
Another important thing to point out is that this program utilizes a few different lists. These lists are initially set up as containers with random or meaningless values. This is intended for the end user to then fill in with their selections. For each of these lists, each respective section of the program will allow to user to modify individual values in accordance to their student selection. This is simply done by taking the index for the student selection, and applying it to the list for the respective variable they wish to change.  

Let's now break down each seciton under the navigation menu.  

## Sections:  

### Student Information Query
We will first begin with the Student Information Query. This section of the program allows the end user to view various bits of information regarding the current students. This section includes values from various lists including the "students" (Students' first and last name), "GPA" (Students' Grade Point Average), Major (What major the students declared, and lastly "violations" (Academic Violations tied to the student). <br>
  
This section works by going through and printing out each respective variable in accordance to the index. To give a better idea of how this works, lets say that currently it is going through and is on student three. The section will print the same index for each of the lists. Think of it like a selection of a row if each of the lists were put together. The program will go through all of the "rows" and will print them all on one line. This section can be of great use in its overview of all of the students in the system.  

When the user is done viewing the information, they may return to the navigation menu by pressing "Y".  

### Student Information Entry
The second option within the navigation menu is the Student Information Entry section. This section is the most complex of the four, but first let's take a quick look at the different options the end user has the ability to change.  

#### New Student
This selection allows the end user to add a new student to the system, providing their first and last name. After the name has been entered, this will then increase the index of the "students" list and input the entered name. NOTE: the original request was to then prompt the user for Major and GPA when a new student is entered, but in consideration of the possibility that the end user may not know this information yet, dummy variables will be entered into the respective lists, and the end user will then be able to go back and enter this information.

#### GPA
This section of the Student Information Query allows the end user to modify the GPA of each individual student. This works by first printing a list of all of the students (including the index) and prompting the end user to then make a selection. Once the student has been selected, the end user will then be prompted to enter in the GPA. When the GPA is entered, the student name and the entered GPA will be printed and a prompt for confirmation will appear. If the user decides to confirm these changes, the entered GPA will then be applied to the selected student. If the end user decides not to confirm, then the changes will be disregarded and then will be returned to the navigation menu.

#### Major
The Major section of the Student Information Entry works very similarly to the GPA section. The user will be prompted to select a student, enter the major, and asked for confirmation. If the user confirms the changes, then the changes will be applied. If they decide not to confirm, the changes will be discarded and the user will then be returned to the navigation menu. 

### GPA Metrics Query
First things first, the original request for the name of this section was "Query GPA Metrics", but the change to GPA Metrics Query was made for consistancy in the naming scheme. In that the Student Information Query, and Query GPA Metrics are not consistant in their titles. With that out of the way, lets go over the simple functionality of this section.  
This section simply grabs the Maximum, Minimum, and Average value from the GPA list. These values are calculated with each instance of selection. Note that if a student has been entered, and the GPA is not changed, the results may be skewed. 

### Academic Violations
Lastly, this section goes over the Avademic Violations. This section allows the end user to select from the list of students, and the list of violations. Firstly the list of students will be printed, then the user will be prompted to select a student. After a student has been selected, the list of possible academic violations will be printed, and the user will select a violation. After both selections have been made, the name of the student and the violation will be printed and the user will be prompted for confirmation. If the user confirms the changes, the changes will be applied. If the user does not confirm the changes, the changes will be disregarded.
